```python
import rarfile
import os
import glob

# Path of uploaded file
rar_path = '/content/Assignment-data.rar'

# Extract folder path
extracted_folder = '/content/Assignment-data'

# Open and extract the file
with rarfile.RarFile(rar_path) as rf:
    rf.extractall(extracted_folder)

print("The file was extracted successfully.")

# Folders
data_path = '/content/Assignment-data/raw_texts'
class_mapping = {'1': 'positive', '2': 'negative', '3': 'neutral'}

all_tweets = []
labels = []
file_names = []

# Files from all folders are retrieved and sorted.
all_files = []
for folder_name, class_label in class_mapping.items():
    folder_path = os.path.join(data_path, folder_name)
    if os.path.exists(folder_path):
        files = glob.glob(os.path.join(folder_path, '*.txt'))
        all_files.extend(files)  # All files are collected in one list

# Sorting files
files_sorted = sorted(all_files, key=lambda x: int(os.path.splitext(os.path.basename(x))[0]))

# Prints sorted file names
print("Sorted file list:")
for file in files_sorted:
    print(os.path.basename(file))

# Reading and processing files
for file in files_sorted:
    try:
        # Opening a file with a specific encoding
        with open(file, 'r', encoding='iso-8859-9') as f:
            tweet = f.read().strip()
            all_tweets.append(tweet)

            # To get class from file name
            if "1" in file:
                class_label = 'positive'
            elif "2" in file:
                class_label = 'negative'
            elif "3" in file:
                class_label = 'neutral'
            labels.append(class_label)

            file_names.append(file)  # The file name is added to the list

    except UnicodeDecodeError as e:
        print(f"An error occurred: {file}, {e}")

print(f"Total {len(all_tweets)} The tweet has been uploaded.")
```

```python
import random
import string
from zemberek.tokenization import TurkishTokenizer
from zemberek.morphology import TurkishMorphology

# Convert to lower case
def to_lowercase(tweet):
    return tweet.lower()

# Tokenization
def tokenize_text(tweet):
    tokenizer = TurkishTokenizer.DEFAULT
    return tokenizer.tokenize(tweet)

# Clear punctuation
def remove_punctuation(tokens):
    return [token for token in tokens if token.content not in string.punctuation]

# Creating a morphology object
morphology = TurkishMorphology.create_with_defaults()

    # Stemming
def apply_stemming(tokens, morphology):
    stemmed_tokens = []
    for token in tokens:
        # If the token contains a hashtag (#) or a special character that cannot be parsed, skip it
        if '#' in token.content or any(char in string.punctuation for char in token.content):
            continue

        analiz = morphology.analyze(token.normalized)
        if analiz.analysis_results:  # Only include words that can be analyzed
            stemmed_tokens.append(analiz.analysis_results[0].item.root)
    return stemmed_tokens



# Random sampling to observe some of the tweets
num_samples = 5
sampled_tweets = random.sample(list(zip(all_tweets, labels)), min(num_samples, len(all_tweets)))

# Applying lower case conversion
print("Step 1 - Convert to Lower Case:")
for i, (tweet, label) in enumerate(sampled_tweets, 1):
    lowercased_tweet = to_lowercase(tweet)
    print(f"{i}. [{label}] {lowercased_tweet[:200]}{'...' if len(lowercased_tweet) > 100 else ''}")

# Implementing tokenization
print("Step 2 - Tokenization:")
for i, (tweet, label) in enumerate(sampled_tweets, 1):
    lowercased_tweet = to_lowercase(tweet)
    tokens = tokenize_text(lowercased_tweet)
    print(f"{i}. [{label}] {[token.content for token in tokens][:20]}{'...' if len(tokens) > 20 else ''}")

# Applying punctuation cleanup
print("Step 3 - Clean Up Punctuation:")
for i, (tweet, label) in enumerate(sampled_tweets, 1):
    lowercased_tweet = to_lowercase(tweet)
    tokens = tokenize_text(lowercased_tweet)
    cleaned_tokens = remove_punctuation(tokens)
    print(f"{i}. [{label}] {[token.content for token in cleaned_tokens][:20]}{'...' if len(cleaned_tokens) > 20 else ''}")

# Applying the stemming process
print("Step 4 - Stemming:")
for i, (tweet, label) in enumerate(sampled_tweets, 1):
    lowercased_tweet = to_lowercase(tweet)
    tokens = tokenize_text(lowercased_tweet)
    cleaned_tokens = remove_punctuation(tokens)
    stemmed_tokens = apply_stemming(cleaned_tokens, morphology)
    print(f"{i}. [{label}] {stemmed_tokens[:20]}{'...' if len(stemmed_tokens) > 20 else ''}")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# The final version of processed tweets can be used directly after stemming.
processed_tweets = []
doc_ids = []  # keep the file numbers

# Create doc_ids_list based on the length of all_tweets
doc_ids_list = list(range(len(all_tweets)))

for tweet, label, doc_id in zip(all_tweets, labels, doc_ids_list):  # doc_ids_list should contain the file number of each tweet
    lowercased_tweet = to_lowercase(tweet)
    tokens = tokenize_text(lowercased_tweet)
    cleaned_tokens = remove_punctuation(tokens)
    stemmed_tokens = apply_stemming(cleaned_tokens, morphology)
    processed_tweets.append(' '.join(stemmed_tokens))  # Final processed tweets
    doc_ids.append(doc_id)  # Include file number for each tweet

# TF-IDF vectorization
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(processed_tweets)

# Creating a DataFrame
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Arrange documents in 'doc_id' order and start from doc1
tfidf_df['class'] = labels
tfidf_df.index = [f"doc{doc_id + 1}" for doc_id in doc_ids]  # Number documents starting from doc1

# Save as CSV
output_path = '/content/tfidf_values.csv'
tfidf_df.to_csv(output_path, sep=';', index=True, encoding='utf-8-sig')
print(f"TF-IDF table '{output_path}' was recorded as.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split,StratifiedKFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
import numpy as np
from statistics import mean, stdev
from sklearn.model_selection import StratifiedKFold

# Load data from CSV file
dataset = pd.read_csv('/content/tfidf_values.csv', sep=';', encoding='utf-8-sig')

# Extract column 'Unnamed: 0' (containing doc1, doc2, etc.) and separate TF-IDF features and classes
X = dataset.drop(columns=['class', 'Unnamed: 0']).to_numpy()  # TF-IDF feature matrix
y = dataset['class'].to_numpy()  # Class labels

# Convert class labels to numeric values using LabelEncoder
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)  # Convert labels to integers

# Step 1: Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.25, random_state=42)

# Step 2: Train KNN Model for k=3
k = 3
classifier = KNeighborsClassifier(n_neighbors=k, metric='cosine')  # 3 nearest neighbors
classifier.fit(X_train, y_train)

# Step 3: Make Predictions on the Test Set
y_pred = classifier.predict(X_test)

# Step 4: Iterate over all test documents and calculate nearest neighbors for k=3
for test_index in range(len(X_test)):
    test_vector = X_test[test_index].reshape(1, -1)  # Test document vector

    print(f"\nTesting document: doc{test_index + 1}")


    # Find nearest neighbors for k=3
    distances, indices = classifier.kneighbors(test_vector)


    # Print results for nearest neighbors and their classes
    for rank, (index, distance) in enumerate(zip(indices[0], distances[0]), start=1):
        print(f"{rank}. doc{index + 1} - Cosine Similarity: {1 - distance:.4f} (Class: {label_encoder.inverse_transform([y_train[index]])[0]})")

    # Majority voting for predicted class (for k=3)
    neighbor_classes = y_train[indices[0]]

    # Majority voting to predict the class
    predicted_class = np.bincount(neighbor_classes).argmax()

    # Print the predicted class based on the value of k
    if k > 1:
        # Print the predicted class for k > 1
        print(f"Estimated class: {label_encoder.inverse_transform([predicted_class])[0]}")
    else:
        # For k = 1, take the closest neighbor's class
        predicted_class = y_train[indices[0][0]]
        print(f"Estimated class: {label_encoder.inverse_transform([predicted_class])[0]}")


# Creating a StratifiedKFold object
n_splits = 10
skf = StratifiedKFold(n_splits=n_splits, shuffle=True, random_state=42)

accuracy_list = []  # List to store the correctness of each fold

# Stratified 10-fold Cross-Validation
for fold, (train_index, test_index) in enumerate(skf.split(X, y_encoded), 1):
    print(f"\nFold {fold}:")

    # Separating training and test sets
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y_encoded[train_index], y_encoded[test_index]

    # Training the model
    classifier.fit(X_train, y_train)

    # Calculating accuracy on test set
    accuracy = classifier.score(X_test, y_test)
    accuracy_list.append(accuracy)
    print(f"Accuracy for Fold {fold}: {accuracy:.4f}")

# Calculate and print performance results
print("\n--- Overall Performance ---")
print(f"List of accuracies: {accuracy_list}")
print(f"Maximum Accuracy: {max(accuracy_list) * 100:.2f}%")
print(f"Minimum Accuracy: {min(accuracy_list) * 100:.2f}%")
print(f"Mean Accuracy: {mean(accuracy_list) * 100:.2f}%")
print(f"Standard Deviation: {stdev(accuracy_list) * 100:.2f}%")
```

```python
import numpy as np

# 1. Determine the k value with the highest average accuracy
best_k = max(results, key=lambda k: results[k]['mean_accuracy'])
print(f"The k value with the highest average accuracy is: {best_k}")

# 2. Train the model for the selected k value and make predictions on the test data.
classifier = KNeighborsClassifier(n_neighbors=best_k, metric='cosine')
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)

# 3. Calculating performance metrics

# a. Accuracy
accuracy = np.mean(y_pred == y_test)
print(f"Accuracy: {accuracy:.4f}")

# b. Calculating the Confusion Matrix
def calculate_confusion_matrix(y_true, y_pred):
    classes = np.unique(y_true)
    cm = np.zeros((len(classes), len(classes)), dtype=int)

    for i, true_class in enumerate(classes):
        for j, predicted_class in enumerate(classes):
            cm[i, j] = np.sum((y_true == true_class) & (y_pred == predicted_class))

    return cm

# Calculate confusion matrix
cm = calculate_confusion_matrix(y_test, y_pred)

# TP, FP, FN, TN
TP = np.diag(cm)
FP = np.sum(cm, axis=0) - TP
FN = np.sum(cm, axis=1) - TP
TN = np.sum(cm) - (TP + FP + FN)

# Calculating Precision and Recall
precision = [TP[i] / (TP[i] + FP[i]) if (TP[i] + FP[i]) > 0 else 0 for i in range(len(TP))]
recall = [TP[i] / (TP[i] + FN[i]) if (TP[i] + FN[i]) > 0 else 0 for i in range(len(TP))]

print(f"Precision for each class:")
for i, p in enumerate(precision):
    print(f"  Class {i+1}: {p:.4f}")

print(f"Recall for each class:")
for i, r in enumerate(recall):
    print(f"  Class {i+1}: {r:.4f}")

# Calculating F1 Score
f1_score = [2 * (precision[i] * recall[i]) / (precision[i] + recall[i]) if (precision[i] + recall[i]) > 0 else 0 for i in range(len(TP))]
print(f"F1 Score for each class:")
for i, f1 in enumerate(f1_score):
    print(f"  Class {i+1}: {f1:.4f}")

#  Printing TP, FP, FN values for each class
print(f"\nTP, FP, FN values for classes:")
for i in range(len(TP)):
    print(f"Class {i+1}:")
    print(f"  Total True Positive (TP): {TP[i]}")
    print(f"  Total False Positive (FP): {FP[i]}")
    print(f"  Total False Negative (FN): {FN[i]}")

#  Macro average and Micro average calculations

# Macro average
macro_precision = np.mean(precision)
macro_recall = np.mean(recall)
macro_f1 = np.mean(f1_score)

print(f"\nMacro Average:")
print(f"  Precision: {macro_precision:.4f}")
print(f"  Recall: {macro_recall:.4f}")
print(f"  F1 Skoru: {macro_f1:.4f}")

# Micro average
micro_TP = np.sum(TP)
micro_FP = np.sum(FP)
micro_FN = np.sum(FN)

micro_precision = micro_TP / (micro_TP + micro_FP) if (micro_TP + micro_FP) > 0 else 0
micro_recall = micro_TP / (micro_TP + micro_FN) if (micro_TP + micro_FN) > 0 else 0
micro_f1 = 2 * (micro_precision * micro_recall) / (micro_precision + micro_recall) if (micro_precision + micro_recall) > 0 else 0

print(f"\nMicro Average:")
print(f"  Precision: {micro_precision:.4f}")
print(f"  Recall: {micro_recall:.4f}")
print(f"  F1 Skoru: {micro_f1:.4f}")
```
